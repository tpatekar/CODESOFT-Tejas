```python
from sklearn.model_selection import train_test_split
x= data[['Votes','Duration','Genre_average','Director_average','Actor1_average','Actor2_average','Actor3_average']]
y=data['Rating']     #target variable
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
x_train, y_train = data[['Votes','Duration','Genre_average','Director_average',
                         'Actor1_average','Actor2_average','Actor3_average']], data['Rating']
```

```python
# to Build a Linear Regression model
model_fit = LinearRegression()
# to Train the model
model_fit.fit(x_train,y_train)
# to Make predictions
predict = model_fit.predict(x_test)
```

```python
# to Evaluating the model's performance
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
MSE =  mean_squared_error(y_test, predict) 
MAE = mean_absolute_error(y_test, predict)
R2_Score = r2_score(y_test, predict)
```

```python
print('Mean squared error value is: ', MSE)
print('Mean absolute error value is: ', MAE)
print('R2 score value is: ', R2_Score)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(y_test, predict, alpha=0.5)
plt.title('Linear Regression Model: Actual vs Predicted Ratings')
plt.xlabel('Actual Ratings')
plt.ylabel('Predicted Ratings')
plt.show()
```
